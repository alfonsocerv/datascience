Son institutos que pueden admitir alumnos de otras partes del distrito escolar, aún fuera de los límites de la zona asignada a la escuela pública; atraen (de ahí la referencia de «imán») a escolares con intereses en general ya definidos.

### Tipos de variables

Las *mediciones* son variables que se pueden cuantificar. La mayoría de las variables anteriores son medidas. Algunas de estas medidas, como state_percentile_16, avg_score_16 y school_rating, son *resultados*.

Los *resultados no* se pueden utilizar para explicarse entre sí. 

Por ejemplo, explicar school_rating como resultado de state_percentile_16 (puntajes de prueba) es una lógica circular. 

     “El argumento (o lógica) circular utiliza su propia conclusión como una de sus premisas 
      establecidas o no. En lugar de ofrecer la prueba, simplemente afirma la conclusión 
      de otra forma, invitando con ello al oyente a aceptarlo como resuelto cuando,
      de hecho, no ha sido resuelta"

La segunda clase, *indicadores*, se utiliza para explicar nuestros resultados.

### Matriz de correlación

La matriz de correlación muestra los valores de correlación de Pearson, que miden el grado de relación lineal entre cada par de elementos o variables. Los valores de correlación se pueden ubicar entre -1 y +1. 

Si los dos elementos tienden a aumentar o disminuir al mismo tiempo, el valor de correlación es positivo, mientras que si la tendencia es opuesta su valor de correlación es negativo. 

Utilice la matriz de correlación para evaluar la fuerza y dirección de la relación entre dos elementos o variables. Un valor de correlación alto y positivo indica que los elementos miden la misma destreza o característica. Si los elementos no están altamente correlacionados, entonces los elementos pudieran medir diferentes características o no estar claramente definidos.

A menudo, las variables con valores de correlación mayores que 0.7 o menores a -0.7 se consideran altamente correlacionadas. Sin embargo, el valor de referencia adecuado que se utilizará también depende de las normas en su área de estudio y el número de elementos en el análisis.

![Ejemplos de matriz de correlación](https://upload.wikimedia.org/wikipedia/commons/3/34/Correlation_coefficient.png)


Python permite evaluar las relaciones lineales entre variables mediante la matrix de correlación ya sea de forma tabular (mediante una tabla) o de forma gráfica utilizando la librería seaborn.

En el diagrama de dispersión anterior, cada punto representa una escuela. La ubicación del punto representa la calificación de esa escuela (eje Y) y el porcentaje de sus estudiantes con almuerzo reducido (eje x).

La línea de tendencia descendente muestra la correlación negativa entre school_rating y reducido_lunch (a medida que uno aumenta, el otro disminuye). La pendiente de la línea de tendencia indica cuánto decrece la calificación escolar a medida que aumenta el almuerzo reducido. Una pendiente más pronunciada indicaría que un pequeño cambio en reducido_lunch tiene un gran impacto en la clasificación_escolar, mientras que una pendiente más horizontal indicaría que el mismo pequeño cambio en reducido_lunch tiene un impacto menor en la clasificación_escuela.

Se evidencia que el diagrama de dispersión respalda aún más lo que vio con el diagrama de caja y bigotes: cuando el almuerzo reducido aumenta, la calificación escolar disminuye. La distribución más estricta de los datos a medida que disminuye la calificación escolar indica la influencia cada vez mayor del almuerzo reducido. Ahora tiene una hipótesis.
